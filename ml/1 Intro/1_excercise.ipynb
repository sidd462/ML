```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot  as plt
#! find canada 2020 earning and draw a regression graph and also plot 2020 earning with a croxx(X) mark on a graph
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
data.head()
```

```python
plt.scatter(data.year,data['per capita income (US$)'])
```

```python
data[['per capita income (US$)']].head()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(data[['year']],data[['per capita income (US$)']])
```

```python
reg.predict([[2020]])       #! Answer
```

```python
plt.scatter(2020,reg.predict([[2020]]), color='blue',marker='x')
```

```python
plt.scatter(data.year,data['per capita income (US$)'])
plt.scatter(2020,reg.predict([[2020]]), color='RED',marker='X')
```

```python
reg.predict(data[['year']])
```
